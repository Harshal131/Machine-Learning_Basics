## One Hot Encoding with multiple categorical features
get_dummies() would work when there are few categorical variables needs to be converted but when there are more categories to be handled then one solution was suggested to resolve this issue in "KDD Cup Orange Challenge". Rather then converting all the variables into dummy variables top 10 most repeated values from the whole column are converted into dummy variables and the rest are set to 0. 

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Datasets/Mercedes_Benz.csv',usecols=['X0', 'X1', 'X2', 'X3', 'X4', 'X5'])
```

```python
df.head()
```

```python
df.X0.value_counts().sort_values(ascending=False).head(10)
```

```python
Label_Name_X0=[x for x in df.X0.value_counts().sort_values(ascending=False).head(10).index]
Label_Name_X0
```

```python
for label in Label_Name_X0:
    df[label]=np.where(df['X0']==label,1,0)
```

```python
df[['X0']+Label_Name_X0].head(10)
```

```python
top_10=[x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]
def one_hot_encoded(data,var,Ten_labels):
    for label in Ten_labels:
        df[label+'_'+var]=np.where(df[var]==label,1,0)
df = pd.read_csv('Datasets/Mercedes_Benz.csv', usecols=['X0', 'X1', 'X2', 'X3', 'X4', 'X5'])
one_hot_encoded(df, 'X1', top_10)
df.head()    
```
