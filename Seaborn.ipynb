## Seaborn 
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
df = sns.load_dataset('iris')
```

## To perform analysis on continuous variables 3 plots can be used:
(1) Distplot
(2) Jointplot 
(3) Pairplot


Note: categorical features wont be considered in the analysis except for pairplot as categorical features can be used as "hue".

## Distplot

Dist plot helps us to perform univariate analysis or to perform distribution of the variables

```python
plt.subplot(2,2,1)
sns.distplot(df['sepal_length']).set_title('Sepal length')
plt.subplot(2,2,2)
sns.distplot(df['sepal_width']).set_title('Sepal width')
plt.subplot(2,2,3)
sns.distplot(df['petal_length']).set_title('Petal length')
plt.subplot(2,2,4)
sns.distplot(df['petal_width']).set_title('Petal width')
plt.tight_layout()
```

```python
# rug plot the datapoints in an array as sticks on an axis
sns.distplot(df['sepal_length'],rug=True,kde=False).set_title('Sepal length')
```

## Joint Plot 
Usually a scatter plot, 2D histogram, 2D density plot used to perform bivariate analysis where the middle part of the graph represents correlation between the variables.

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
```

```python
#This function will regress y on x (possibly as a robust or polynomial regression) and then draw a scatterplot of the residuals. You can optionally fit a lowess smoother to the residual plot, which can help in determining if there is structure to the residuals.
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='resid')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='scatter')
```

## Pair Plot 
Pairplot creates a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.

```python
sns.pairplot(df,hue='species')
```

```python
sns.pairplot(df,hue='species',kind='reg')
```

## Analysis of categorical features can be performed by:
(1) Boxplot
(2) Violinplot
(3) Countplot
(4) Barplot
(5) Pointplot

## Boxplot

```python
sns.boxplot(x='sepal_length',y='sepal_width',data=df)
```

## Violinplot 

```python
sns.violinplot(x='sepal_length',y='sepal_width',hue='species',data=df)
```

## Countplot 

```python
sns.countplot('species',data=df)
```

## Barplot 

```python
sns.barplot(x='sepal_length',y='species',data=df)
```

## Pointplot 

```python
sns.pointplot(x='sepal_length',y='sepal_width', hue="species", kind="point", data=df);
```
